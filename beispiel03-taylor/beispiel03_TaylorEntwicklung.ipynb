# Programmbeispiel: Taylorentwicklung

Das folgende Beispiel benötigt die 'symbolic' Toolbox von Octave [https://octave.sourceforge.io/symbolic/]. Diese kann über folgenden Befehl installiert:

```octave
pkg install -forge symbolic
```

Sobald die Toolbox einmal installiert ist, muss sie beim nächsten Mal nur noch geladen werden:

```octave
pkg load symbolic
```

Nun können wir mit symbolischen Variablen arbeiten:

```octave
syms x
```

sowie Ableitungen berechnen (vgl. [Dokumentation zu](https://octave.sourceforge.io/symbolic/function/@sym/diff.html) `diff`): 

```octave
f = exp(x*x)
diff(f,x)
```

Auch Taylorreihen werden unterstützt (vgl. [Dokumentation zu](https://octave.sourceforge.io/symbolic/function/@sym/taylor.html) `taylor`)

```octave
f = exp(x)
T0 = taylor(f, x, 0, 'order', 1)
T1 = taylor(f, x, 0, 'order', 2)
T2 = taylor(f, x, 0, 'order', 3)
T3 = taylor(f, x, 0, 'order', 4)
T4 = taylor(f, x, 0, 'order', 5)
T5 = taylor(f, x, 0, 'order', 6)
```

Symbolische Funktionen können mit `ezplot` dargestellt werden:

```octave
%plot -f svg
hold on
ezplot(f)
ezplot(T0)
ezplot(T1)
ezplot(T2)
% ezplot(T3)
% ezplot(T4)
xlim([-3 3])
ylim([-4 10])
legend({"f", "T_0(f)", "T_1(f)", "T_2(f)", "T_3(f)", "T_4(f)"})
grid on
```

# Weitere Beispiele:

$f(x) = x^3, a = 1$

```octave
f(x) = x*x*x;
a = sym(1);
Tf_0=taylor(f, x, a, 'order', 1);
Tf_1=taylor(f, x, a, 'order', 2)
Tf_2=taylor(f, x, a, 'order', 3)
Tf_3=taylor(f, x, a, 'order', 4)

newplot()
hold on
ezplot(f)
ezplot(Tf_0)
ezplot(Tf_1)
ezplot(Tf_2)
ezplot(Tf_3)
xlim([-1 3])
ylim([-5 15])
legend({"f", "T_0(f)", "T_1(f)", "T_2(f)", "T_3(f)"})
grid on
```

$f(x) = \sin(x), a = \frac \pi 2$

```octave
f(x) = sin(x)
a = sym(pi)/2;
f(a)

Tf_0=taylor(f, x, a, 'order', 1)
Tf_1=taylor(f, x, a, 'order', 2)

Tf_2=taylor(f, x, a, 'order', 3)
Tf_3=taylor(f, x, a, 'order', 4)

Tf_4=taylor(f, x, a, 'order', 5)
% Tf_5=taylor(f, x, a, 'order', 6)

Tf_6=taylor(f, x, a, 'order', 7)

newplot()
hold on
ezplot(f)
ezplot(Tf_0)
ezplot(Tf_2)
ezplot(Tf_4)
ezplot(Tf_6)
xlim([-4 8])
ylim([-4 4])
legend({"f", "T_0(f)", "T_2(f)", "T_4(f)", "T_6(f)"})
grid on
```
