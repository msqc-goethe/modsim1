```octave
n=80
x=linspace (-1.0, 1.0, n)
h=2.0/(n-1)
[X,Y]=meshgrid(x,x);
```

```octave
Room=(((X>-1) & (X<1) & (Y>-0.5) & (Y< 0.5)) | ((X>-0.5) & (X<0.5) &  (Y<0.5) & (Y>-1.0)) )
```

```octave
# Heater=((X>-0.3) & (X<0.3) & (Y>-0.75) & (Y<-0.5)) # Heizung am Fenster
Heater=((X>-0.3) & (X<0.3) & (Y>-0.3) & (Y<0.3))  # Heizung mittig
#Heater=((X>0.5) & (X<0.75) & (Y>-0.3) & (Y<0.3))  # Heizung rechts
```

```octave
cells = find(Room)
```

```octave
Room = zeros(size(Room))
Room(cells) = (1:length(cells))'
mesh(X,Y,Room)
```

```octave
HeaterInd = Room(find(Heater>0))
WindowInd=Room(2,find(Room(2,:)>0))(:)
```

```octave
# Indizes fuer die Tuer
DoorInd = Room(:,2)
f=find(DoorInd>0)
x(f)
DoorInd = DoorInd(find(DoorInd>0))
d = length(DoorInd)
d/2
DoorInd = DoorInd(d/2-d/4:d/2+d/4)
```

```octave
function A=create_laplacian(G, n)
    nnz = max(max(G))
    A =zeros(nnz);
    for i=2:n-1
        for j=2:n-1
             ind = G(i,j);
             left = G(i-1,j);
             right = G(i+1,j);
             top = G(i,j+1);
             bot = G(i,j-1);
             
             if (ind == 0) continue; endif
             A(ind,ind) = 4.0;
             if (bot > 0) A(ind,bot) = -1.0; endif
             if (top > 0) A(ind,top) = -1.0; endif
             if (left > 0) A(ind,left) = -1.0; endif
             if (right > 0) A(ind,right) = -1.0; endif
        endfor
    endfor
end
```

```octave
function A=set_neumann(G, n, A)
    for i=2:n-1
        for j=2:n-1
             ind = G(i,j);
             left = G(i-1,j);
             right = G(i+1,j);
             top = G(i,j+1);
             bot = G(i,j-1);
             
             if (ind == 0) continue; endif
            
             if (bot == 0) A(ind,ind) -= 1.0; endif
             if (top == 0) A(ind,ind) -= 1.0; endif
             if (left == 0) A(ind,ind) -= 1.0; endif
             if (right == 0) A(ind,ind) -= 1.0; endif
        endfor
    endfor
end
```

```octave
function [A,b] = set_dirichlet(i, ui, A, b)
    b(i) = ui;
    A(i,:) = 0;
    A(i,i) = eye(length(i));
endfunction
```

```octave
A=create_laplacian(Room, n);
A=set_neumann(Room, n, A);
A/=(h*h);
```

```octave

N = max(size(A));
b=zeros(N,1);
HeaterInd
b(HeaterInd) = 100.0; # 100.0

# Dirichlet values
DoorTemp = 15.0;
WindowTemp = 5.0;

DoorInd
[A,b]=set_dirichlet(DoorInd, DoorTemp, A, b);
WindowInd
[A,b]=set_dirichlet(WindowInd, WindowTemp, A, b);
```

```octave

u=linsolve(A,b);

U=Room;
U(Room>0)=full(u(Room(Room>0)));
mesh(X,Y,U)
```
